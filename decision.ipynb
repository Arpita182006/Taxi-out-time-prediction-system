```python
#import libraries
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
#reading dataset

ds=pd.read_csv('cleaned_dat.csv')
ds=ds.loc[:,~ds.columns.str.contains('^unnamed')]
ds
```

```python
ds.info()
```

```python
#label encoding on categoricol data

from sklearn.preprocessing import LabelEncoder
le_airline = LabelEncoder()
le_weather = LabelEncoder()

ds['Airline'] = le_airline.fit_transform(ds['Airline'].str.strip().str.lower())
ds['Weather'] = le_weather.fit_transform(ds['Weather'].str.strip().str.lower())
```

```python
ds.columns
```

```python
# Split into features and target

x= ds[['Departure_Hour', 'Flight_Distance_km',
       'Departing_Flights', 'Airline', 'Runways_Available', 'Weather',
       'Aircraft_Weight_kg']]
y = ds['Taxi_Out_Time']
```

```python
x
```

```python
y
```

```python
#train test split 
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score

# Initialize the model
dt_model = DecisionTreeRegressor(random_state=42)

# Train the model
dt_model.fit(x_train, y_train)

# Make predictions
y_pred = dt_model.predict(x_test)

# Evaluate
print("Decision Tree Regression Results:")
print("R2 Score:", r2_score(y_test, y_pred))

```

```python
import pandas as pd

# Take user input (use numeric values for categorical fields)
dep_hour = int(input("Enter Departure Hour (0â€“23): "))
distance = float(input("Enter Flight Distance (km): "))
departing = int(input("Enter Departing Flights: "))
airline = int(input("Enter Airline (0 = airline_a, 1 = airline_b): "))
runways = int(input("Enter Runways Available: "))
weather = int(input("Enter Weather (0 = clear, 1 = rainy): "))
weight = int(input("Enter Aircraft Weight (kg): "))

# Create DataFrame
cols = ['Departure_Hour', 'Flight_Distance_km', 'Departing_Flights',
        'Airline', 'Runways_Available', 'Weather', 'Aircraft_Weight_kg']

user_input = pd.DataFrame([[dep_hour, distance, departing, airline, runways, weather, weight]],
                          columns=cols)

# Predict
result = dt_model.predict(user_input)
print("Estimated Taxi-Out Time (in minutes):", round(result[0], 2))
```
