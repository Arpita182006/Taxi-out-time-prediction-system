```python
#importing libraries

import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
ds=pd.read_csv('taxi_ou_uncleaned_data.csv')
ds
```

```python
ds.info()
```

```python
#handling missing values

cat_cols = ['Airline',  'Weather']
for col in cat_cols:
    ds[col] = ds[col].fillna(ds[col].mode()[0])
```

```python
#for numeric column

num_col=['Departure_Hour','Flight_Distance_km','Departing_Flights','Runways_Available','Aircraft_Weight_kg','Taxi_Out_Time']
for cols in num_col:
    ds[cols]=ds[cols].fillna(ds[cols].mean())
```

```python
print(ds.isnull().sum())
```

```python
# Clean Airline and Weather columns
ds['Airline'] = ds['Airline'].str.strip().str.lower()
ds['Weather'] = ds['Weather'].str.strip().str.lower()

# Optionally map to standardized names
ds['Airline'] = ds['Airline'].replace({
    'airline a': 'airline_a',
    'airline_b': 'airline_b'
})

ds['Weather'] = ds['Weather'].replace({
    'clear': 'clear',
    'rainy': 'rainy'
})
```

```python
ds.to_csv('cleaned_dat.csv')
```
